```python
import pandas as pd
import numpy as np
df = pd.read_csv('Self_employment_rate.csv')

year = df.FREQUENCY.unique()
print(len(year) , year)

df.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def draw(country):
    
    d1 = df.loc[(df['LOCATION'] == country)]
    x = d1['TIME']
    y = d1['Value']
    x = [str(i) for i in x]
    plt.figure(figsize=(10,5))
    plt.xlabel('YEAR', fontsize = 20)
    plt.ylabel('VALUE', fontsize = 20)
    plt.title(country, fontsize = 50)
    plt.bar(x, y, label = country, width = 0.4)
    plt.legend()
    plt.xticks(rotation = 90)
    plt.show()
```

```python
countries = df.LOCATION.unique()
print(len(countries))
print(countries)
for country in countries:
      draw(country)
      
```
